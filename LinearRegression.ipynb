```python
def hypothesis(x,theta):
    return theta[0]+theta[1]*x
```

```python
def error(x,y,theta):
    error = 0
    for i in range(x.shape[0]):
        hx=hypothesis(x[i],theta)
        error += (hx-y[i])**2
    return error    
```

```python
def gradient(x,y,theta):
    grad=np.zeros((2,1))
    
    for i in range(x.shape[0]):
        hx=hypothesis(x[i],theta)
        
        grad[0]+= (hx-y[i])
        grad[1]+= (hx-y[i])*x[i]
    return grad    
    
```

```python
def gradientdescent(x,y,lr=0.0001):
    theta=np.array([0,0])
    error_list=[]
    theta_list=[]
    
    for i in range(100):
        grad= gradient(x,y,theta)
        err=error(x,y,theta)
        error_list.append(err)
        theta_list.append(theta)
        
        theta[0]-=lr*grad[0]
        theta[1]-=lr*grad[1]
        
    return theta_list,error_list,theta
```

```python
theta_list,error_list,final_theta = gradientdescent(x,y)
```
